```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris_datset = datasets.load_iris()
print(iris_datset)
```

```python
X = iris_datset.data
print(X.shape)
Y = iris_datset.target
print(Y.shape)
```

```python
pca = PCA(2)
pca
```

```python
pca.fit(X)
```

```python
pca.components_
```

```python
Z = pca.transform(X)
Z.shape
```

```python
plt.scatter(Z[:,0], Z[:,1], c=Y);
```
