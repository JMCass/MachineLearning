```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rng = np.random.RandomState(0)
rng
```

```python
W = rng.randn(1000,2)
W
```

```python
plt.scatter(W[:,0], W[:,1]);
plt.axis('equal')
plt.grid('on')
```

```python
A = rng.rand(2,2)
A
```

```python
X = np.dot(W, A)
X
```

```python
plt.scatter(X[:,0], X[:,1]);
plt.axis('equal')
plt.grid('on')
```

```python
np.var(X[:,0])
```

```python
np.var(X[:,1])
```

```python
covar = np.cov(X.T)
covar
```

```python
e, V = np.linalg.eig((covar))
```

```python
e
```

```python
V
```

```python
V[:,0]
```

```python
V[:,1]
```

```python
u = V[:,1]
u.shape
```

```python
Z = np.dot(X, u)
Z.shape
```

```python
np.var(Z)
```

```python
np.dot(u,u)
```

```python
Y = np.outer(Z,u)
Y.shape
```

```python
plt.scatter(X[:,0], X[:,1])
plt.scatter(Y[:,0], Y[:,1])
plt.axis('equal')
plt.grid('on');
```
